```python
import cv2
import matplotlib.pyplot as plt
import  numpy as np
```

# **1. Read Image**

```python
image_path = "WhatsApp Image 2024-10-11 at 19.45.33_a892c801.jpg"
image = cv2.imread(image_path)
plt.imshow(image)
plt.show()
```

# **2. Convert image to HSV color space**

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
plt.show()
```

```python
# Define the range of blue color in HSV (adjust the lower and upper bounds if needed)
lower_blue = np.array([90, 50, 50])   # Lower bound for blue
upper_blue = np.array([130, 255, 255])  # Upper bound for blue
```

# **3. Creating mask for the blue color**

```python
mask = cv2.inRange(hsv, lower_blue, upper_blue)
```

# **4. Finding contours from the mask**

```python
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

# **5. Create a black background image**

```python
filled_image = np.zeros_like(image)
```

# **6. Fill the detected object (contours) with white**

```python
cv2.drawContours(filled_image, contours, -1, (255, 255, 255), thickness=cv2.FILLED)
```

# **7. Concatenate the original and the filled image horizontally**

```python
concatenated_image = np.hstack((image, filled_image))
```

# **8. Display Concatenated Image**

```python
plt.figure(figsize = (12, 6))
plt.imshow(cv2.cvtColor(concatenated_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image & Detected Object (White)")
plt.axis("off")
plt.show()
```
